```python
import pandas as pd
import numpy as np
from sklearn.model_selection import LeaveOneGroupOut
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
import joblib
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
def RF_model():
    param_grid = {'n_estimators': [5, 10, 20, 40, 60, 80, 100],
                  'max_features': [2, 4, 6, 8, 10],
                  'max_depth': [2, 4, 6, 8, 10],
                 }
#    param_grid = {['n_estimators': [5, 10, 20, 40], 'max_depth': [2, 4, 6, 8], 
#                   'min_samples_split': 5, 'max_features': [2, 3, 4],
#                   'learning_rate': 0.01, 'loss': 'squared_error',]}
#'learning_rate': [0.01, 0.02, 0.05, 0.1],
    forest_reg = GradientBoostingRegressor()
    grid_search = RandomizedSearchCV(forest_reg, param_grid, cv=15, 
                               scoring='squared_error',
                               return_train_score=True, n_iter=50)
    return grid_search
```

```python
# Load data
cs_file = '../data/SMB_input_all_ERA5.csv'
#cs_file = '../data/SMB_input_four_ERA5.csv'

df = pd.read_csv(cs_file,
   delimiter='\t', index_col=['Date'],
    parse_dates=['Date'], na_values='NAN')
labels = df.drop_duplicates(subset=['Stake'])

loso = LeaveOneGroupOut()
groups = df['Stake'].values

X = (df.drop(['SMB', 'Stake'], axis=1)).to_numpy()
y = (df['SMB'].copy()).to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1000)
```

```python
model = RF_model().fit(X_train, y_train)
```

```python
model.best_params_
```

```python
test_scores1 = model.cv_results_['mean_test_score']
train_scores1 = model.cv_results_['mean_train_score'] 

fig, (ax0) = plt.subplots(figsize=(6,3))
ax0.plot(test_scores1, label='test')
ax0.plot(train_scores1, label='train')
ax0.set_ylabel('Score') #fontsize=25
ax0.set_xlabel('n_estimators') #fontsize=25
ax0.legend(loc='best')

grid_ytest_p = model.predict(X_test)
grid_test_mse = mean_squared_error(y_test, grid_ytest_p)
grid_test_rmse = np.sqrt(grid_test_mse)
print(grid_test_rmse)
```
